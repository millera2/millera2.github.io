## **TRIANGLE**

First: drawing a triangle with color.

```python
import turtle

oscar = turtle.Turtle()   #name of turtle

def makeTriangle(myTurtle, shape, x, y, color, sideLength, heading):
    """this will make a triangle using these given inputs.
    our output will be the shape this def makes"""
    myTurtle.setheading(heading)
    myTurtle.penup()
    myTurtle.goto(x,y)
    myTurtle.pendown()
    myTurtle.pencolor(color)
    myTurtle.fillcolor(color)
    myTurtle.begin_fill()
    for j in range(3):     #because we are now dealing w a triangle, range would be 3 (3 sides)
        myTurtle.forward(100)
        myTurtle.left(120) #angle is 120 and not 60 because it is going left i.e. 180-60=120
    myTurtle.end_fill()
    
def main():
    """this will make the code run correctly"""
    oscar = turtle.Turtle()
    makeTriangle(oscar, 'triangle', 100, 100, 'LightSeaGreen', 100, 0)
    #main function is where we use the arguments for the makeTriangle() 
    #always have to make sure to put define the main()
    
    
main()
    
#always remember to return the main() so the code runs
    
    
    
```
