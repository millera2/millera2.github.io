import turtle

def makeSquare(userColor, userHeading, userX, userY, userSideLength):
    """
    Draws A Square
    
    Parameters:
         userColor: Color string that the user chooses
         userHeading:A string that is converted to a numerical value, setting the turtle's heading
         userX: A string that is converted to a numerical value, setting the turtle's x position
         userY: A string that is converted to a numerical value, setting the turtle's y position
         userSideLength: Sets the Length of each segment of the square

              None
    """
    
    
    
    
    tortoise = turtle.Turtle() #allows the turtle to make illustrations
    tortoise.pencolor("black") #sets tortoise's pencolor to black
    tortoise.fillcolor(userColor) #sets fillcolor to user's input
    tortoise.heading() #relays current heading
    tortoise.position() #relays current position
    tortoise.up() #lifts tortoise's tail so he can move if need be without drawing
    tortoise.setheading(userHeading) #sets heading to user's input
    tortoise.goto(userX,userY) #sends tortoise to user's x and y input
    tortoise.down() #sets tortoise's tail down so he can start drawing
    tortoise.begin_fill() #begins fill of color
    for segment in range(4): #draws a square
        tortoise.forward(userSideLength)
        tortoise.left(90)
    tortoise.end_fill()
