```julia
using Plots
using LaTeXStrings

function D2_1(n)
    D = zeros(n,n)

    for i in 1:n
        if i==1
            D[i,:] = [-1 1 zeros(n-2)']
        elseif i==n
            D[i,:] = [zeros(n-1)' 1 ]
        else
            D[i,:] = [zeros(i-2)' 1 -2 1 zeros(n-1-i)']
        end
    end
    
    return D
end

function D2_2(n)
    D = zeros(n,n)

    for i in 1:n
        if i==1
            D[i,:] = [-3 4 -1 zeros(n-3)']
        elseif i==n
            D[i,:] = [zeros(n-1)' 1 ]
        else
            D[i,:] = [zeros(i-2)' 1 -2 1 zeros(n-1-i)']
        end
    end
    
    return D
end

function g1(x)
    return(cos(π*x/2))
end

function u1(x)
    return ( (4/π^2)*cos(π*x/2) )
end





"This function plots the original u, and the interpolates based on its second derivative
Input:
    u = function to be interpolated
    g = second derivative of u, ie u''(x) = g(x).
"
function plotInterpolates(u, g, n)

    
    h = 1/n
    
    xvals = LinRange(0,1,n+1)      # The x-vals where g(x) is known
    
    
    image = (-h^2).*[g(x) for x in xvals]   #find the "data vector on RHS of 1.27, 1.29"
    image[1] = 0
    image[n+1] = 0

    
    
    plot(xvals, inv(D2_1(n+1))*image     , label = "first order", bg=false )
    plot!(xvals, inv(D2_2(n+1))*image    , label = "second order" )
    plot!( LinRange(0,1,100), [u(x) for x in LinRange(0,1,100)], label= "u")    #plot u
    
end
```

```julia
plotInterpolates(x->1-x^8, x->8*7x^6, 4)
```
