```julia
using RDatasets

iris = dataset("datasets", "iris")
head(iris)
```

```julia
using Gadfly

Gadfly.plot(iris, x=:SepalLength, 
            y=:SepalWidth, 
            color=:Species, 
            Geom.point,
            Theme(discrete_highlight_color=c->nothing))

```
