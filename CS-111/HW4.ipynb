```python
import matplotlib.pyplot as pyplot

def getTotals(myList):
    total = 0
    totals = []
    for value in myList:
        total = total+value
        totals.append(total)
    return totals





def monthlyPayment(p, rate, payment):
    principal = p
    interest = p*rate/12
    newPrincipal = principal - (payment-interest)
    return [newPrincipal, interest, payment-interest]


def paymentHistory(closing, down, r, payment):
  
    principal = closing-down                      # 
    principalPayments = [0]              #   Initialize values
    interestPayment = [0]                #  
    balances = [principal]               #
    
    while(principal >0):
        
        results = monthlyPayment(principal, r, payment)     #<------- most important line!!!
        
        principalPayment = results[2]                   #   
        interest = results[1]                           #   Update values
        principal = results[0]                          #
        
        principalPayments.append(principalPayment)     #
        interestPayment.append(interest)               #   add to accumulators
        balances.append(principal)                     #
        
    return [principalPayments, interestPayment, balances]



def printRecords(principals, interests, balances):
    formatString = '{0:^9}|{1:^11.2f}|{2:^11.2f}|{3:^9.2f}'
    for i in range(len(principals)):
        print(formatString.format(str(i//12)+"y "+str(i%12)+"m ", balances[i], interests[i], principals[i]))
       


def plotTotals(principals, interests, balances):
    
    months = list( range( len(principals) ) )
    pyplot.plot(months, getTotals(principals))
    pyplot.plot(months, getTotals(interests))
    pyplot.plot(months, balances)
    
    fs = "{0:^10.2f}"
    print("Total Interest Paid")
    print(fs.format(getTotals(interests)[-1] ))
        
        
    
   

    
```

```python
results = paymentHistory(165000, 0, .05, 950)


plotTotals(results[0], results[1], results[2])
```
